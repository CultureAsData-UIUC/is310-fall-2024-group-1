```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import pandas as pd
df = pd.read_csv("IS310 DATA - Sheet1.csv")
df
```

```python
df_pre_2011 = df[df['YOR'] < 2011]
df_post_2011 = df[df['YOR'] >= 2011]

X_pre_2011 = df_pre_2011[['YOR']]
y_pre_2011 = df_pre_2011['NO. OF SKINTONES']

model_pre_2011 = LinearRegression()
model_pre_2011.fit(X_pre_2011, y_pre_2011)

df_pre_2011['predicted_skin_tones'] = model_pre_2011.predict(X_pre_2011)

X_post_2011 = df_post_2011[['YOR']]
y_post_2011 = df_post_2011['NO. OF SKINTONES']

model_post_2011 = LinearRegression()
model_post_2011.fit(X_post_2011, y_post_2011)

df_post_2011['predicted_skin_tones'] = model_post_2011.predict(X_post_2011)
df_combined = pd.concat([df_pre_2011, df_post_2011])
df_combined[['game_name', 'YOR', 'NO. OF SKINTONES', 'predicted_skin_tones']]
```

```python
mae_pre_2011 = mean_absolute_error(df_pre_2011['NO. OF SKINTONES'], df_pre_2011['predicted_skin_tones'])
mae_post_2011 = mean_absolute_error(df_post_2011['NO. OF SKINTONES'], df_post_2011['predicted_skin_tones'])
metrics = {
    "Pre-2011 Model": {
        "MAE": mae_pre_2011,
    },
    "Post-2011 Model": {
        "MAE": mae_post_2011,
    }
}

metrics_df = pd.DataFrame(metrics).T
metrics_df
```

Measuring Accuracy:

This is the average of the absolute differences between the predicted values and the actual values. It measures the magnitude of errors in a set of predictions, without considering their direction (positive or negative).

Interpretation:

Pre-2011 Model: The MAE is 1.83, meaning that, on average, the predicted number of skin tones is off by about 1.83 skin tones from the actual values.
Post-2011 Model: The MAE is 4.15, which means the predictions are, on average, off by about 4.15 skin tones.

1. Understanding Trends Over Time:

Pre-2011 vs. Post-2011 Trends: By analyzing the regression models for pre-2011 and post-2011 games, we can see how skin tone diversity has shifted over the years. The analysis shows that games from the pre-2011 era generally had fewer skin tones, with predictions for the post-2011 period showing an increase in diversity. This supports the idea that societal shifts toward inclusivity and diversity likely influenced game developers to offer more skin tones in later years.

2. Eliminating Manual Data Entry:

The prediction model allows us to instantly predict the number of skin tones for any game, based on its year of release. This speeds up the entire data collection process, especially when dealing with large datasets or numerous games. For instance, if we were previously analyzing hundreds of games, manually counting skin tones for each would take considerable time. The model allows for quicker analysis and the ability to process large volumes of data in much less time.
